```python
import numpy as np
import pandas as pd

class LinearRegression():

    def __init__(self, learning_rate, iterations):
        self.learning_rate = learning_rate
        self.iterations = iterations

    def fit(self, X, Y):
        self.m, self.n = X.shape          # m — кол-во примеров, n — кол-во признаков
        self.weights = np.zeros(self.n)   # веса
        self.bias = 0                     # свободный член
        self.X = X
        self.Y = Y

        for i in range(self.iterations):
            self.update_weights()

        return self

    def update_weights(self):
        Y_pred = self.predict(self.X)
        error = Y_pred - self.Y                              # вектор ошибок

        dW = (1 / self.m) * self.X.T.dot(error)             # градиент по весам
        db = (1 / self.m) * np.sum(error)                   # градиент по bias

        self.weights -= self.learning_rate * dW
        self.bias    -= self.learning_rate * db

        return self

    def predict(self, X):
        return X.dot(self.weights) + self.bias


# Загрузка данных и обучение
df = pd.read_csv('salary_data.csv')

X = df.iloc[:, :-1].values   # признаки (YearsExperience)
Y = df.iloc[:, 1].values     # целевая переменная (Salary)

model = LinearRegression(iterations=1000, learning_rate=0.01)
model.fit(X, Y)

Y_pred = model.predict(X)

print(*np.round(model.weights, 2))
print(np.round(model.bias, 2))
```
